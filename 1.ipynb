```python
import pandas as pd
import numpy as np

df=pd.read_csv("C:/Users/PALLE  BALU/OneDrive/Desktop/ch/marksheet.csv")
df
```

```python
total=df['Maths'] + df['Science'] + df['English']
per= (total/400)*100
df['percentage']=per
df
```

```python
df['result']=np.where(df['percentage']>=30,'pass','fail')
df
```

```python
count=df.groupby(['Gender','result']).size()
count.plot(kind='pie')
```

```python
sec=df.groupby(['Section','result']).size()
sec
```

```python
l=df.groupby('Section')['result'].apply(
    lambda x: (x=='pass').mean() *100
)
l
```

```python
l.plot(kind='bar')
```
